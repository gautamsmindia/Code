```r
data <- read.csv("d:\\alldata\\kyphosis.csv")
```

```r
str(data)
```

```r
data$Kyphosis <- as.factor(data$Kyphosis)
```

```r
head(data)
```

```r
library(caTools)
```

```r
splits <- sample.split(data$Kyphosis,SplitRatio=0.7)
```

```r
train <- subset(data,splits==TRUE)
test <- subset(data,splits==FALSE)
```

```r
model <- glm(Kyphosis~.,data=data,family = "binomial")
```

```r
model
```

```r
dim(train)
```

```r
tail(test)
```

```r
predictions <- predict(model,test)
```

```r
labels<-ifelse(predictions>0.5,"present","absent")
```

```r
mean(test$Kyphosis==labels)
```
